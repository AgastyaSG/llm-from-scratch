```python
#attention , 3.3.1

#create mini context vector 
#dot product
#softmax
#
import torch
inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

```python
query = inputs[1]
att_score_2=[0.0]*6
att_score_2 =torch.tensor( [torch.dot(x_i, query) for i,x_i in enumerate(inputs)])
#comprehension syntax

print(att_score_2)
```

```python
att_wt_2=torch.softmax(att_score_2,0) #dim??, softmax better than simple sum and divide, calculate for large val
torch.set_printoptions(precision=10, sci_mode=False)
print(att_wt_2)
```

```python
# 3.1.2
att_scores = inputs @ inputs.T
print(att_scores)
```

```python
att_wts = torch.softmax(att_scores, dim=-1)
print(att_wts)
```

```python
#3.3.2.3
all_context = att_wts @ inputs
print(all_context)
```
