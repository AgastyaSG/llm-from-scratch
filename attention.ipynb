```python
class SelfAttention_v2(nn.Module):
    def __init__(self, d_in, d_out, qkv_bias=False):
        super().__init__()
        self.W_query = nn.Linear(d_in, d_out, bias=qkv_bias)
        self.W_key   = nn.Linear(d_in, d_out, bias=qkv_bias)
        self.W_value = nn.Linear(d_in, d_out, bias=qkv_bias)
        #print(f"W_q= {self.W_query}'\n', W_k = { self.W_key }'\n', W_v = {self.W_value }")
    def forward(self, x):
        #A = torch.eye(3, 2)
        print(x.shape)
        
        keys = self.W_key(x)
        queries = self.W_query(x)
        values = self.W_value(x)
        attn_scores = queries @ keys.T
        attn_weights = torch.softmax(
            attn_scores / keys.shape[-1]**0.5, dim=-1
        )
        context_vec = attn_weights @ values
        return context_vec
```

```python
class CausalAttention(nn.Module):
    def __init__(self,d_in,d_out,context_length, unknown,qkv_bias = False):
        super().__init__()
        self.W_keys = torch.nn.Linear(d_in, d_out, bias = qkv_bias )
        self.W_queries = torch.nn.Linear(d_in, d_out, bias = qkv_bias)
        self.W_values = torch.nn.Linear(d_in, d_out, bias = qkv_bias)
        self.dropout = nn.Dropout(0.5,True) #self.dropout??, dropout nature
        self.register_buffer(
           'mask',
           torch.triu(torch.ones(context_length, context_length),
           diagonal=1)
        ) 
        
    def forward(self,x):
        print(f"W_values = {self.W_values}, x shape = {x.shape}")
        keys = self.W_keys(x)
        values = self.W_values(x)
        queries =  self.W_queries(x)
        attn_scores = keys @ queries.transpose(1,2)
        #causal mask !
        #-inf
        mask = torch.triu(torch.ones(x.shape),diagonal = 1)
        attn_scores.masked_fill(mask.bool(),-torch.inf)
        attn_wts = torch.softmax(attn_scores / keys.shape[-1]** 0.5,dim = -1)
        
        attn_wts = self.dropout(attn_scores)
        
        context_vec = attn_wts @ values
        return context_vec
```
